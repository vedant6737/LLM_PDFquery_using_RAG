```python
#Split text using Character Text Splitter
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=800,
    chunk_overlap=200,
    length_function=len
)

texts = text_splitter.split_text(raw_text)

print("Number of chunks:", len(texts))
```

```python
retriever = astra_vector_store.as_retriever(
    search_kwargs={"k": 3}
)
```

```python
query = "What is the document about?"

docs = retriever.invoke(query)

for i, doc in enumerate(docs):
    print(f"\nDocument {i+1}")
    print(doc.page_content[:500])
```

```python
first_question = True

while True:

    if first_question:
        query_text = input("\nEnter your question (or type 'quit' to exit): ").strip()
    else:
        query_text = input("\nWhat's your next question (or type 'quit' to exit): ").strip()

    if query_text.lower() == "quit":
        break

    if query_text == "":
        continue

    first_question = False

    print(f'\nQUESTION: "{query_text}"')

    # Retrieve relevant chunks
    docs = astra_vector_store.similarity_search(
        query_text,
        k=3
    )

    # Combine retrieved text
    context = "\n\n".join([doc.page_content for doc in docs])

    # Ask LLM
    prompt = f"""
    Answer the question using only the context below.

    Context:
    {context}

    Question:
    {query_text}
    """

    response = llm.invoke(prompt)

    print("\nANSWER:")
    print(response.content)

    print("\nFIRST DOCUMENTS BY RELEVANCE:")

    docs_with_scores = astra_vector_store.similarity_search_with_score(
        query_text,
        k=3
    )

    for doc, score in docs_with_scores:
        print(f"\n[{score:.4f}] | {doc.page_content[:100]}...")
```
